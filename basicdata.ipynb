### Banda de ondas classificadas no EEG

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

time = np.arange(0, 2, 0.001);

delta_waves = np.sin(0.3*np.pi*time) + np.sin(1*np.pi*time) + np.sin(4*np.pi*time)
theta_waves = np.sin(4.5*np.pi*time) + np.sin(6*np.pi*time) + np.sin(8*np.pi*time)
alpha_waves = np.sin(8.5*np.pi*time) + np.sin(10*np.pi*time) + np.sin(13*np.pi*time)
beta_waves = np.sin(14*np.pi*time) + np.sin(20*np.pi*time) + np.sin(25*np.pi*time) + np.sin(30*np.pi*time)
lowgamma_waves = np.sin(31*np.pi*time) + np.sin(40*np.pi*time) + np.sin(45*np.pi*time) + np.sin(50*np.pi*time) + np.sin(80*np.pi*time)
highgamma_waves = np.sin(80.5*np.pi*time) + np.sin(90*np.pi*time) + np.sin(100*np.pi*time) + np.sin(125*np.pi*time) + np.sin(150*np.pi*time)

figure = plt.figure(figsize=(15,25))

deltax = figure.add_subplot(611)
deltax.set_title('Delta Waves')
deltax.set_xlim(left=0, right=2)
deltax.xaxis.set_major_locator(ticker.NullLocator())
deltax.yaxis.set_major_locator(ticker.NullLocator())
deltax.plot(time, delta_waves)

thetax = figure.add_subplot(612)
thetax.set_title('Theta Waves')
thetax.set_xlim(left=0, right=2)
thetax.xaxis.set_major_locator(ticker.NullLocator())
thetax.yaxis.set_major_locator(ticker.NullLocator())
thetax.plot(time, theta_waves)

alphax = figure.add_subplot(613)
alphax.set_title('Alpha Waves')
alphax.set_xlim(left=0, right=2)
alphax.xaxis.set_major_locator(ticker.NullLocator())
alphax.yaxis.set_major_locator(ticker.NullLocator())
alphax.plot(time, alpha_waves)

betax = figure.add_subplot(614)
betax.set_title('Beta Waves')
betax.set_xlim(left=0, right=2)
betax.xaxis.set_major_locator(ticker.NullLocator())
betax.yaxis.set_major_locator(ticker.NullLocator())
betax.plot(time, beta_waves)

lowgammax = figure.add_subplot(615)
lowgammax.set_title('Low Gamma Waves')
lowgammax.xaxis.set_major_locator(ticker.NullLocator())
lowgammax.yaxis.set_major_locator(ticker.NullLocator())
lowgammax.set_xlim(left=0, right=2)
lowgammax.plot(time, lowgamma_waves)

highgammax = figure.add_subplot(616)
highgammax.set_title('High Gamma Waves')
highgammax.xaxis.set_major_locator(ticker.NullLocator())
highgammax.yaxis.set_major_locator(ticker.NullLocator())
highgammax.set_xlim(left=0, right=2)
highgammax.plot(time, highgamma_waves)

plt.show();
```

### Transformação domínio do tempo para frequência

```python
import numpy as np
import matplotlib.pyplot as plt

N = 5000
T = 1.0 / 100.0
time = np.linspace(0.0, N*T, N)
frequency = np.linspace(0.0, 1.0/(2.0*T), N//2)

##Frequencies: 0.3Hz && 1Hz && 4Hz
temporal_wave = np.sin(0.3*2*np.pi*time) + np.sin(2*np.pi*time) + np.sin(8*np.pi*time)
frequency_wave = np.fft.fft(temporal_wave)

figure = plt.figure(figsize=(5,5))

temporalx = figure.add_subplot(211)
temporalx.set_title('Temporal Waves')
temporalx.set_xlim(left=0, right=2)
temporalx.xaxis.set_major_locator(ticker.NullLocator())
temporalx.yaxis.set_major_locator(ticker.NullLocator())
temporalx.plot(time, temporal_wave)

frequencyx = figure.add_subplot(212)
frequencyx.set_title('Frequency Bands')
frequencyx.set_xlim(left=0, right=6)
frequencyx.plot(frequency, 2.0/N*np.abs(frequency_wave[:N//2]) )
```
